Checking for Outliers

```python
outliers=df[df['Sales Count']>1000]
if len(outliers)>0:
    print(outliers[['Timestamp','Sales Count','Redemption Count']])
else:
    print("No extreme outliers found")
```

Acitivity week by week

```python
df['DayOfWeek']=df['Timestamp'].dt.day_name()
day_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
weekly=df.groupby('DayOfWeek')['Sales Count'].sum().reindex(day_order)
for day,sales in weekly.items():
    print(f" {day}: {sales:,}")
```

Acitivity by Hour

```python
df['Hour']=df['Timestamp'].dt.hour
hourly=df.groupby('Hour')['Sales Count'].sum()
peak_hour=hourly.idxmax()
print(f"\n Peak Hour: {peak_hour}:00 ({hourly[peak_hour]:,} tickets)")
```

Weekday vs Weekend

```python
df['IsWeekend']=df['DayOfWeek'].isin(['Saturday','Sunday'])
weekend_sales=df[df['IsWeekend']]['Sales Count'].sum()
weekday_sales=df[~df['IsWeekend']]['Sales Count'].sum()
print(f"\nWeekend Sales: {weekend_sales:,}")
print(f"\nWeekday Sales: {weekday_sales:,}")
print(f"Weekday-Weekend Ratio: {weekend_sales/weekday_sales:.1f}x")
```
